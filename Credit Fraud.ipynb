```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
df['Time_scaled'] = scaler.fit_transform(df[['Time']])
df['Amount_scaled'] = scaler.fit_transform(df[['Amount']])

df.drop('Time', axis = 1, inplace = True)
df.drop('Amount', axis = 1, inplace = True)
```

```python
df[['Time_scaled', 'Amount_scaled']]
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('Class', axis = 1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression

logModel = LogisticRegression(max_iter = 500)

logModel.fit(X_train, y_train)

y_pred = logModel.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn import metrics
def model_stats(y_test, y_pred):
  print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
  print("Precision: ", metrics.precision_score(y_test, y_pred))
  print("Recall: ", metrics.recall_score(y_test, y_pred))
```

```python
model_stats(y_test, y_pred)
```

Going to utilize oversampling and see if that improves anything

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state = 42)

X_train_os, y_train_os = smote.fit_resample(X_train, y_train)
print(pd.value_counts(y_train))
print(pd.value_counts(y_train_os))
```

```python
logModel.fit(X_train_os, y_train_os)

y_pred = logModel.predict(X_test)

model_stats(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred, labels = logModel.classes_)

cm_disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = ['not fraud', 'fraud'])

cm_disp.plot()
```

Oversampling increase our recall from 0.6 to 0.92. However, our precision droped from 0.85 to 0.05. I can also conclude that oversampling helped the model better identify fraud, but it is overestimating for fraud. Considering the problem at hand, it is better for there to be false positives than false negatives. Therefore it is okay for precision to be lower than our recall. However, there could be the arugment that it could annoy customers if they constantly checked if there was fraud even though they were all good.

I am going to tune the model and see if that helps anything.

```python
from sklearn.model_selection import GridSearchCV

new_log = LogisticRegression()

param_grid = [
    {
        'penalty' : ['l1' , 'l2' , 'elasticnet', 'None'],
        'C' : [0.1, 1, 10, 100, 1000],
        'solver' : ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
        'max_iter' : [500]
    }
]

gridsearch = GridSearchCV(new_log, param_grid = param_grid, cv = 3, n_jobs = -1, verbose = 4)

gridsearch.fit(X_train_os, y_train_os)
```

```python
print(gridsearch.best_params_)
new_log = gridsearch.best_estimator_
```

```python
y_pred = new_log.predict(X_test)
model_stats(y_test, y_pred)
```

Going to go for undersampling approach

```python
nonfraud = df.loc[df['Class'] == 1]
fraud = df.loc[df['Class'] == 0].sample(n = 492)
```

```python
undersample_df = pd.concat([nonfraud, fraud], axis=0)
```

```python
undersample_df
```

```python
#us stands for undersample
X_us = undersample_df.drop('Class', axis = 1)
y_us = undersample_df['Class']
```

```python
X_train_us, X_test_us, y_train_us, y_test_us = train_test_split(X_us, y_us, random_state = 42, stratify = y_us)
```

```python
pd.value_counts(y_train)
```

```python
logModel.fit(X_train_us, y_train_us)

y_pred = logModel.predict(X_test_us)
```

```python
model_stats(y_test_us, y_pred)
```

Undersampling greatly increased precision on the test set

```python
cm = confusion_matrix(y_test_us, y_pred, labels = logModel.classes_)

cm_disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = ['not fraud', 'fraud'])

cm_disp.plot()
```
